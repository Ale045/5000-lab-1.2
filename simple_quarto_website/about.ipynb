```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)  
y = np.sin(x)  


plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)', color='blue')
plt.title('Sine Curve using Matplotlib')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.legend()
plt.show()
```
