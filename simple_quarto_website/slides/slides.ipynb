## Python Plot Demo

```python

import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 400)

# Compute the corresponding y values
y = x**2

# Plot the function
plt.plot(x, y, label="y = x^2")
plt.title("Plot of y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
